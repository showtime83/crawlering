```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://mania.kr/"
```

```python
page = requests.get(url)
```

```python
page_cont = page.content
```

```python
page_bs = BeautifulSoup(page_cont, "html.parser")
```

```python
page_bs
```

```python
section_board = page_bs.find("div", {"class" : "bottom_row_2col"})
```

```python
board_cont = section_board.find_all("div", {"class" : "cut_str rc_box_list_2col_best_a_inner_team"})
```

```python
board_cont
```

```python
title_list = []
for i in board_cont :
    title_list.append(i.text.strip())
```
